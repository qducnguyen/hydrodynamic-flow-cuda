# Import library

```python
import numpy as np
from matplotlib import pyplot, cm
```

# Reading results file 



```python
RESULT_FILE_PATH = "flow_results.txt"
```

```python
with open(RESULT_FILE_PATH) as f:
    results = f.readlines()

nx = int(results[0].strip())
ny = int(results[1].strip())
xmax = float(results[5].strip())
ymax = float(results[6].strip())


u = np.zeros((ny, nx))
v = np.zeros((ny, nx))
p = np.zeros((ny, nx))

start = 12
for i in range(ny):
    result = results[start+i].strip().split() 
    for j in range(nx):
        u[i][j] = float(result[j])

start = nx + 12
for i in range(ny):
    result = results[start+i].strip().split() 
    for j in range(nx):
        v[i][j] = float(result[j])

start = 2*nx + 12
for i in range(ny):
    result = results[start+i].strip().split() 
    for j in range(nx):
        p[i][j] = float(result[j])

```

# Visualisation

```python
x = np.linspace(0, xmax, nx)
y = np.linspace(0, ymax, ny)

X, Y = np.meshgrid(x, y)

fig = pyplot.figure(figsize=(11,7), dpi=100)
# plotting the pressure field as a contour
pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
pyplot.colorbar()
# plotting the pressure field outlines
pyplot.contour(X, Y, p, cmap=cm.viridis)
# plotting velocity field
pyplot.quiver(X[::2, ::2], Y[::2, ::2], u[::2, ::2], v[::2, ::2])
pyplot.xlabel('X')
pyplot.ylabel('Y');
```

```python
fig = pyplot.figure(figsize=(11, 7), dpi=100)
pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
pyplot.colorbar()
pyplot.contour(X, Y, p, cmap=cm.viridis)
pyplot.streamplot(X, Y, u, v)
pyplot.xlabel('X')
pyplot.ylabel('Y');
```
